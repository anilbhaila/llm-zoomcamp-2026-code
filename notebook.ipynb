```python
import requests

docs_url = 'https://datatalks.club/faq/json/courses.json'
response = requests.get(docs_url)
courses_raw = response.json()

documents = []
url_prefix = 'https://datatalks.club/faq'

for course in courses_raw:
    course_url = f'{url_prefix}{course['path']}'

    course_response = requests.get(course_url)
    course_response.raise_for_status()
    course_data = course_response.json()

    documents.extend(course_data)

len(documents)
```

```python
documents[100]
```

```python
from minsearch import Index

index = Index(
    text_fields=['question', 'section', 'answer'],
    keyword_fields=['course']
)

index.fit(documents)
```

```python
def search(question, course='llm-zoomcamp'):
    boost_dict={'question': 2.0, 'section':0.5}
    filter_dict={'course': course}
        
    return index.search(
        question,
        boost_dict=boost_dict,
        filter_dict=filter_dict,
        num_results=5
    )
```

```python
search_results = search(question)
search_results
```

```python
INSTRUCTIONS = '''
Your task is to answer questions from the course participants based on the provided context.

Use the context to find relevant information and provide accurate answers. If the answer is not found in the context, respond with "I don't know."
'''
```

```python
USER_PROMPT_TEMPLATE = '''
Question:
{question}

Context:
{context}
'''
```

```python
def build_context(search_results):
    lines = []

    for doc in search_results:
        lines.append(doc['section'])
        lines.append('Q: ' + doc['question'])
        lines.append('A: ' + doc['answer'])
        lines.append('')
    
    return '\n'.join(lines).strip()
```

```python
context = build_context(search_results)
print(context)
```

```python
def build_prompt(question, search_results):
    context = build_context(search_results)
    prompt = USER_PROMPT_TEMPLATE.format(
        question=question, 
        context=context
    )
    return prompt.strip()
```

```python
prompt = build_prompt(question,search_results)
print(prompt)
```

```python
response = openai_client.responses.create(
    model='gpt-5.4-mini',
    input=prompt
)
```

```python
input_price = 0.75 / 1000000
output_price = 4.50 / 1000000

cost = (
    response.usage.input_tokens * input_price +
    response.usage.output_tokens * output_price
)

cost
```

```python
def llm(instructions, user_prompt, model='gpt-5.4-mini'):
    message_history = [
        {'role':'developer', 'content': instructions},
        {'role':'user', 'content':user_prompt}
    ]

    response = openai_client.responses.create(
        model= model,
        input=message_history
    )
    
    return response.output_text
```

```python
res = llm(INSTRUCTIONS, prompt)
print(res)
```

```python
def rag(query, model='gpt-5.4-mini'):
    search_reasults = search(query)
    prompt = build_prompt(query, search_reasults)
    answer = llm(INSTRUCIONS,prompt, model=model)
    return answer
```

```python
answer = rag(question)
print(answer)
```
